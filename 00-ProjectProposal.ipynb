## Research Question

How do the current state policies against e-cigarette, nicotine, and tobacco products across the United States impact the amount of vape usage among youth, as measured by factors such as self-reported vape usage, quit attempts, and social media sentiment about vaping? 

## Background and Prior Work

Our project aims to establish a correlation between state policies and the usage of vapes and e-cigarettes among youth by assessing factors such as peer influence, marketing tactics, and industry attempts to bypass state regulations and restrictions. To conduct this research, we plan to compare the percentage of vape usage per state, the percentage of those attempting to quit, and the overall sentiment among youth in comparison to the severity or leniency of their state policies and enforcement. Since there is more data specific to each county in California, we will also investigate the relationship between income levels, quality of education, and accessibility of vape devices among specific age groups to determine what influences a child to begin vaping at such a young age. According to the Nation Youth Tobacco Survey, there was a drastic 1000% increase in the amount of high school students who used disposable vapes from 2019 to 2020. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) 

A prominent reason why children and teens are more inclined to try vaping products is because they are easily accessible and concealable. Despite FDA regulations intended to discourage teens and young adults from vaping, current policies only target specific devices and flavors, making it easy for the industry to circumvent these restrictions by inventing new flavors, new names, and new products that are not yet regulated in order to continue selling nicotine products to kids. What’s more, children and teens can easily purchase these products online or in stores, where age verification is often not required. As of now, young people are more likely to become addicted to nicotine through vaping rather than smoking cigarettes, a major shift from previous generations. 
    
Smoking and vaping has been a prevalent topic among both children and young adults as cigarettes become less appealing to the younger generation. Instead, vapes are being marketed by the industry as a safer and fun alternative to cigarettes, often advertised alongside fun fruity flavors and small, easily concealable devices. As a result, the Youth Risk Behavior survey shows that the percentage of teenagers who smoke cigarettes have significantly dropped from “70% in 1991 down to 58.4% in 2003 and even further down to 28.9% in 2017”(Jones, Salzman, par 2).<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)  Given this data, it’s easy to assume that the increase in anti-smoking advocacy and decades long research on cigarettes has successfully influenced teenagers and young adults to quit smoking. 
    
However, these figures only show one side of the story. With the invention of new e-cigarettes and vape products, a study published by Stanford University found that out of the 4351 participants,  2167 of them used vapes and e-cigarettes, with some participants being as young as thirteen years old. That’s almost 50% of participants that smoke, with two thirds still under the minimum legal age of 21. As more youth are gaining access to vapes and e-cigarettes, there is a higher risk of forming life long habits and addiction to nicotine. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) 
    
In recent years, the number of youths reported using e-cigarettes have increased with 1.63 million reporting usage in 2025. Vaping habits can be influenced through different social and demographic factors. According to the American Lung Association (ALA), those who have an undergraduate degree have a smoking rate of 4.4% compared to those with a high school degree with the rate of 14.7% and 18.3% for those who did not graduate high school. However, education isn’t the only factor that may contribute to increased smoking rates. Household income has the same effect. Youth in high income families are 28% more likely to view vaping as harmful in comparison to those from lower income families. <a name="cite_ref-9"></a>[<sup>9</sup>](#cite_note-9)
    
As for health risks, smoking cigarettes has been studied extensively but vaping is relatively new. In fact, the e-cigarettes that are currently available on the U.S. market have not been systematically reviewed by the Food and Drug Administration (FDA) for their long-term effects on lung health. Neither of the two is considered safe, but since there are a lot of unknown factors in vaping, people turn a blind eye. Public health organizations such as the American Lung Association have expressed serious concern that e-cigarettes pose significant risks to respiratory health. 
        
Both smoking and vaping pose serious health risks as they expose users to harmful chemicals that can cause irreversible lung damage and diseases. In 2018, the National Academies of Sciences, Engineering, and Medicine published a consensus report that reviewed more than 800 studies on e-cigarettes and their effects. This report concluded that e-cigarette use is associated with clear health risks and that these products both contain and emit potentially toxic substances. The report also found moderate evidence that youth who use e-cigarettes experience increased coughing, wheezing, and asthma exacerbations. This challenges the perception that vaping is harmless. Additional research shows that common e-cigarette ingredients include propylene glycol and vegetable glycerin. These substances are toxic to cells; not only that, but e-cigarettes produce harmful chemicals such as formaldehyde, acetaldehyde, and acrolein, which are associated with lung and heart disease. <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)
    
Additionally, most cigarettes and e-cigarettes contain nicotine, a highly addictive substance that is especially damaging for young adults and adolescents whose brains are still developing. Nicotine use in youth can negatively affect attention, learning, mood and impulse control, and mental health, leading to heightened stress, depressive symptoms, and suicidal ideation (Patten Par 6). <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)
    
Exposure risks extend beyond direct use. Both the U.S. Surgeon General and the National Academies have warned that secondhand e-cigarette emissions contain nicotine, ultrafine particles, volatile organic compounds, heavy metals, and flavoring chemicals such as diacetyl that have been linked to serious lung disease. These findings suggest that vaping can pose health risks not only to users but also to those exposed indirectly. 
    
What makes vaping so enticing is the way it's marketed to teens and young adults. Advertisements with bright colors or fun visuals are quick to draw in a younger audience. An increase in social media is also helping in widespread advertising of nicotine products. According to the research paper Exposure to e-cigarette advertising and young people's use of e-cigarettes: A four country study the “majority (85%) of respondents who had heard of e-cigarettes reported being exposed to e-cigarette advertising on at least one type of media” (Jones, Santos, et al). <a name="cite_ref-10"></a>[<sup>10</sup>](#cite_note-10) Even with certain restrictions in countries, the access of social media and vape shops continues to expose young people to nicotine. In a similar paper titled The vaping epidemic in adolescents, advertisements of fun flavors such as “mint to gummi bear to frosted sugar cookie” (Jones, Salzman par 5)<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)  , are made to target younger audiences. They say that flavor is the most important factor and catering to their tastes with new flavors is the primary reason they try vapes. On top of that new vapes come in cool packaging and can be customised to your liking, some including screens on the sides. Widespread media and targeted flavors and aesthetics are a leading cause in the increase of teen vaping.

## Hypothesis


We hypothesize that states with more restrictive e-cigarette and tobacco policies will have lower rates of youth vape usage and higher rates of quit attempts among youth. We also expect that states with stricter policies will exhibit more negative sentiment toward vaping on social media platforms compared to states with more lenient regulations.


## Data

To answer this question, we would need data about e-cigarette, nicotine, and tobacco policies across the United States as well as data about e-cigarette, nicotine, and tobacco usage among youth in the United States. The variables that we would need about policies are: whether or not each state has laws restricting e-cigarette use, whether or not each state has laws restricting nicotine use, and whether or not each state has laws restricting tobacco use. More specifically, we would like to have variables about policies regarding youth, such as the legal age to buy e-cigarettes, nicotine, and tobacco in each state. Regarding usage, the variables we would need are survey data such as: a person’s age, the state they live in, whether or not they use e-cigarettes, nicotine, and tobacco, how often they use e-cigarettes, nicotine, and tobacco each week, if they are planning on quitting these substances, if they have already quit these substances, and a score representing their general attitude towards e-cigarettes, nicotine, and tobacco.

Ideally, we would like to have at least 1000 observations for a complete analysis. The data would be collected from online databases such as the National Youth Tobacco Survey, state policy databases, and other survey data found through the internet. The data will be downloaded in .csv format, and read into pandas dataframes for analysis.

The first potential dataset that we found is linked on the website: https://www.publichealthlawcenter.org/resources/us-e-cigarette-regulations-50-state-review. This dataset provides information about e-cigarette policies in all 50 US states. It is stored in pdf format, so we would have to convert it to csv format manually in order to use it. The important variables in this dataset are whether or not each state has a tax on e-cigarettes, whether or not each state has laws regulating e-cigarette packaging, whether or not each state has youth access restrictions, and whether or not each state requires a retail license to sell e-cigarettes.

The second potential dataset that we found is located at https://www.kaggle.com/datasets/mikhail1681/youth-tobacco-survey-yts-data-usa. This dataset contains data from the Youth Tobacco Survey that was given to middle and high school students regarding tobacco use. It is already in .csv format and can be used directly after downloading. The important variables in this dataset are a person’s age, state, and use status (current, frequent, or ever) for cigarettes and smokeless tobacco products.

The third potential dataset that we found is located at https://www.kaggle.com/datasets/bestgirl/ots-emotions2. This dataset contains tweets about vaping from 2015 to 2022. It is in .csv format and can be downloaded directly. The important variables in this dataset are the content of the tweet, the date and time that the tweet was posted at, the state that the tweet was posted in, and the sentiment class of each tweet. Data like this could help us determine the overall sentiment about e-cigarettes in each state.
