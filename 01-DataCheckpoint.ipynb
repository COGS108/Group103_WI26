## Background and Prior Work

 Youth vaping has become a significant health issue over the last two decades, one that has been made more complicated due to differing state policies across the country. To gain a better understanding, our project aims to establish a correlation between state regulations and the usage of vapes and e-cigarettes among youth, ages 11 to 18, per state by assessing factors such as self-reported vape usage, quit attempts, and social media sentiment about vaping. By doing so, our data seeks to uncover whether the enforcement of these regulations contribute to overall reduction in vape usage. 
 
Our background addresses the serious health implications of vaping, challenging the widespread misconception that it is harmless and non-addictive. To do so, we will examine national usage trends to highlight the scope of youth vaping and analyze the influence of industry marketing and advertising, in addition to the role of education and income in shaping tobacco use patterns. Lastly, we will review prior research to identify the gap that our study addressed. 

There are a lot of health risks associated with smoking cigarettes that have been studied extensively but vaping is relatively new. In fact, the e-cigarettes that are currently available on the U.S. market have not been systematically reviewed by the Food and Drug Administration (FDA) for their long-term effects on lung health. Neither of the two is considered safe, but since there are a lot of unknown factors in vaping, people turn a blind eye. Public health organizations such as the American Lung Association have expressed serious concern that e-cigarettes pose significant risks to respiratory health. 

Both smoking and vaping pose serious health risks as they expose users to harmful chemicals that can cause irreversible lung damage and diseases. In 2018, the National Academies of Sciences, Engineering, and Medicine published a consensus report that reviewed more than 800 studies on e-cigarettes and their effects. This report concluded that e-cigarette use is associated with clear health risks and that these products both contain and emit potentially toxic substances. The report also found moderate evidence that youth who use e-cigarettes experience increased coughing, wheezing, and asthma exacerbations.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

Additionally, most cigarettes and e-cigarettes contain nicotine, a highly addictive substance that is especially damaging for young adults and adolescents whose brains are still developing. Nicotine use in youth can negatively affect attention, learning, mood and impulse control, and mental health, leading to heightened stress, depressive symptoms, and suicidal ideation (Patten Par 6). <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)

Exposure risks extend beyond direct use. Both the U.S. Surgeon General and the National Academies have warned that secondhand e-cigarette emissions contain nicotine, ultrafine particles, volatile organic compounds, heavy metals, and flavoring chemicals such as diacetyl that have been linked to serious lung disease. These findings suggest that vaping can pose health risks not only to users but also to those exposed indirectly. 

According to the Nation Youth Tobacco Survey, there was a drastic 1000% increase in the amount of high school students who used disposable vapes from 2019 to 2020 <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). A prominent reason why children and teens are more inclined to try vaping products is because they are easily accessible and concealable. Despite FDA regulations intended to discourage teens and young adults from vaping, current policies only target specific devices and flavors, making it easy for the industry to circumvent these restrictions by inventing new flavors, new names, and new products that are not yet regulated in order to continue selling nicotine products to kids. What’s more, children and teens can easily purchase these products online or in stores, where age verification is often not required. As of now, young people are more likely to become addicted to nicotine through vaping rather than smoking cigarettes, a major shift from previous generations.

 Smoking and vaping has been a prevalent topic among both children and young adults as cigarettes become less appealing to the younger generation. Instead, vapes are being marketed by the industry as a safer and fun alternative to cigarettes, often advertised alongside fun fruity flavors and small, easily concealable devices. As a result, the Youth Risk Behavior survey shows that the percentage of teenagers who smoke cigarettes have significantly dropped from “70% in 1991 down to 58.4% in 2003 and even further down to 28.9% in 2017”(Jones, Salzman, par 2).<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Given this data, it’s easy to assume that the increase in anti-smoking advocacy and decades long research on cigarettes has successfully influenced teenagers and young adults to quit smoking. 
 
However, these figures only show one side of the story. With the invention of new e-cigarettes and vape products, a study published by Stanford University found that out of the 4351 participants,  2167 of them used vapes and e-cigarettes, with some participants being as young as thirteen years old. That’s almost 50% of participants that smoke, with two thirds still under the minimum legal age of 21. As more youth are gaining access to vapes and e-cigarettes, there is a higher risk of forming life long habits and addiction to nicotine. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

What makes vaping so enticing is the way it's marketed to teens and young adults. Advertisements with bright colors or fun visuals are quick to draw in a younger audience. An increase in social media is also helping in widespread advertising of nicotine products. According to the research paper Exposure to e-cigarette advertising and young people's use of e-cigarettes: A four country study the “majority (85%) of respondents who had heard of e-cigarettes reported being exposed to e-cigarette advertising on at least one type of media” (Jones, Santos, et al).<a name="cite_ref-10"></a>[<sup>10</sup>](#cite_note-10) Even with certain restrictions in different states, the access of social media and vape shops continues to expose young people to nicotine. In a similar paper titled The vaping epidemic in adolescents, advertisements of fun flavors such as “mint to gummi bear to frosted sugar cookie” (Jones, Salzman par 5),<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) are made to target younger audiences. Teens say that flavor is the most important factor and catering to their tastes with new flavors is the primary reason they try vapes. On top of that new vapes come in cool packaging and can be customised to your liking, some including screens on the sides. Widespread media and targeted flavors and aesthetics are a leading cause in the increase of teen vaping. 

In recent years, the number of youths reported using e-cigarettes have increased with 1.63 million reporting usage in 2025. Vaping habits can be influenced through different social and demographic factors. According to the American Lung Association (ALA), those who have an undergraduate degree have a smoking rate of 4.4% compared to those with a high school degree with the rate of 14.7% and 18.3% for those who did not graduate high school. However, education isn’t the only factor that may contribute to increased smoking rates. Household income has the same effect. Youth in high income families are 28% more likely to view vaping as harmful in comparison to those from lower income families.  <a name="cite_ref-9"></a>[<sup>9</sup>](#cite_note-9)

Overall, the studies described above primarily focus on adult participants from a specific state. When they do address vaping among youth, the analysis is often concentrated on only one factor that impacts youth vaping rates. Our project aims to encompass teenage usage, from between the ages of 11 to 18, across all states to gain a nationwide overview. We also plan to look at a variety of regulations that impact youth vaping rates, like retail license regulations, age restrictions, and more. By doing so, we may uncover the reason behind early vape usage and implement better preventative measures to reduce the overall usage, a pressing concern that past studies have often overlooked.
