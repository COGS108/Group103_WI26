## Abstract

Youth vaping has become an increasing public health concern in the United States, especially amongst teenagers. This project looks into whether there is a correlation between strict state vaping policies and lower rates of youth vaping. We analyzed data ranging from 2000 to 2017 across all states in the U.S., concentrating on policies such as retail license requirements, minimum age for purchasing vape products, marketing regulations, and restrictions near protected public spaces. We used geographic visualizations, exploratory data analysis, an ANOVA test, regression analysis, and k-means clustering to examine the correlation between state policy strengths, rates of youth vaping, and public sentiment about vaping on Twitter. 

Through these analyses, it was found that states with less regulations around the usage and sale of vape products had higher rates of youth vaping. This relationship suggests that regulations and laws have a significant impact on behavior, specifically regarding children and the products that they have access to. However, no relationship was found between state regulations and public sentiment regarding vaping on Twitter. The impact of laws and regulations about vape usage is not reflected on social media in the same way that it is in youth behavior. Collectively, these findings emphasize the impact that state policies have on youth vaping rates and social media sentiment about vaping.

## Hypothesis


We hypothesize that states with stronger tobacco control policies, including higher minimum legal purchase ages, mandatory retail license requirements, stricter advertising regulations, and restrictions on sales near protected public spaces, will have lower rates of self-reported youth vaping and higher rates of quit attempts among individuals aged 11 to 18 between 2000 and 2017. We also expect that states with stronger regulatory environments will demonstrate more negative social media sentiment toward vaping compared to states with more lenient policies. Reduced availability and visibility of these products may decrease initiation and increase efforts to stop use. ### Dataset #3 - Vape and Juul Tweet data 
The dataset, “Vape and JUUL Dataset (Tweet Location),” contains 30,000 publicly available tweets collected from Twitter to examine sentiment toward vaping and e-cigarettes. The dataset includes 15 variables and primarily consists of English-language tweets posted after 2015 using keywords such as “juul,” “juuls,” and “vaping.” Each row represents a single tweet and includes information such as the tweet content, username, date and time posted, sentiment score, and user location.   

The variables most relevant to this project are the sentiment score, timestamp, and state location. The sentiment score is a continuous numeric variable ranging from approximately -0.99 to 1. Values closer to -1 indicate strongly negative sentiment toward vaping, values near 0 indicate neutral sentiment, and values closer to 1 indicate strongly positive sentiment. The date and time of the tweet allow us to examine changes in sentiment over time, while the location variable allows us to compare sentiment across states with different regulatory environments. Like Dataset #2, this dataset was already compiled and structured, and we did not manually input or reconstruct the data.   

This dataset is useful because it provides insight into public attitudes toward vaping and allows us to examine whether sentiment differs across states. However, there are several limitations. ### Analysis #1 - ANOVA

To statistically evaluate whether youth vaping rates differ across policy strength levels, we conducted a one-way Analysis of Variance (ANOVA). This test compares the mean youth vaping rates across the different policy score categories to determine whether the differences observed in the boxplot are statistically significant rather than occurring by chance.

```python
from scipy.stats import f_oneway

# group youth vaping rates by policy category
groups = [group["avg_youth_use"].dropna()
    for name, group in combined_data.groupby("policy_score_cat")]

# run ANOVA
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

The results of the one-way ANOVA indicate that there are statistically significant differences in average youth vaping rates across different policy strength levels (F = 4.90, p = 0.003). Since the p-value is well below the conventional significance threshold of 0.05, we reject the null hypothesis that all policy groups have the same average youth vaping rate. This suggests that the strength of vaping-related policies is associated with differences in youth vaping prevalence across states. Combined with the earlier visualizations and correlation analysis, these results provide additional evidence that stronger policy environments may be linked to lower youth vaping rates.

### Analysis #2 - Regression

The next type of analysis that we ran on the data was ordinary least squares regression. A regression analysis quantifies the effect of policies on average youth vape use. In this analysis, the dependent variable is the average youth vaping rate in each state, while the independent variable is the policy strength score. This approach allows us to evaluate whether stricter policies are predictive of lower youth vaping rates.

```python
import statsmodels.api as sm

reg_data = combined_data[["avg_youth_use", "policy_score"]].dropna()
X = reg_data[["policy_score"]]
y = reg_data["avg_youth_use"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

The regression results show that policy strength has a statistically significant negative relationship with youth vaping rates. The coefficient for policy_score is -1.33 (p = 0.004), indicating that every time the policy strength score increases by one, the average youth vaping rate decreases by about 1.33. This suggests that states with stricter regulatory environments tend to have lower levels of youth vaping.

The overall regression model is statistically significant (F = 9.64, p = 0.0036), which means that policy strength is able to explain the differences that exist in youth vaping rates across states. However, the model’s R² value of 0.202 indicates that policy strength only accounts for about 20% of the variation in youth vaping rates. This suggests that while policy differences have an impact on youth vaping rates, other factors likely also influence youth vaping behavior.

Overall, these results reinforce the findings from the earlier ANOVA analysis. Both analyses suggest that stronger policy environments are associated with lower youth vaping rates, providing evidence that state-level regulations may help reduce youth vaping prevalence.

### Analysis #3 - K-Means Clustering

For the final analysis, we implemented K-means clustering. K-means clustering works to identify natural groupings of states based on policy strength, youth vaping rates, and tweet sentiment. Before clustering, all variables were standardized to ensure that differences in scale did not bias the results

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

cluster_data = combined_data[["policy_score", "avg_youth_use", "avg_sentiment"]].dropna()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(scaled_data)
cluster_data["cluster"] = cluster_labels

centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("Cluster Centers:\n[Policy Score   Vaping Rate   Sentiment]\n", centers)
```

In the cluster centers shown above, the first column represents policy score, the second column represents youth vaping rates, and the third column represents tweet sentiment. As you can see, the first cluster center (first row) has the highest policy score and the lowest youth vaping rate. The last cluster center (bottom row) has the lowest policy score and the highest youth vaping rate. The middle cluster center (second row) has a moderate policy score and a moderate youth vaping rate. The sentiment scores do not vary much across clusters.

Since K-means clustering is an unsupervised algorithm, it finds natural patterns in the data without using pre-existing labels. The fact that K-means was able to cleanly seperate states into high-policy / low vaping rate states and low-policy / high vaping rate states shows that there is an underlying relationship between policy strength and youth vaping behavior in the data. Overall, stronger vaping policies do correlate with lower youth vaping rates in each state. At the same time, twitter sentiment did not appear to change much between clusters, further showing that social media attitudes are not strongly impacted by state regulations. Both of these results reinforce the findings from earlier analyses.
