## Abstract
Youth vaping has become an increasing public health concern in the United States, especially amongst teenagers. This project looks into whether there is a correlation between strict state vaping policies and lower rates of youth vaping. We analyzed data ranging from 2000 to 2017 across all states in the U.S., concentrating on policies such as retail license requirements, minimum age for purchasing vape products, marketing regulations, and restrictions near protected public spaces. We used geographic visualizations, correlation analysis, exploratory data analysis, and an ANOVA test to examine the correlation between state policy strengths, rates of youth vaping, and public sentiment about vaping on Twitter. 

Through these analyses, it was found that states with less regulations around the usage and sale of vape products had higher rates of youth vaping. This relationship suggests that regulations and laws have a significant impact on behavior, specifically regarding children and the products that they have access to. However, no relationship was found between state regulations and public sentiment regarding vaping on Twitter. The impact of laws and regulations about vape usage is not reflected on social media in the same way that it is in youth behavior. Collectively, these findings emphasize the impact that state policies have on youth vaping rates and social media sentiment about vaping.

It includes students in grades 6–8 and 9–12, and each observation represents aggregated results for a specific state in a given year. The dataset contains 10,600 observations and 31 variables. The most relevant variables for our project include the state, year, age or grade group, response category, and data value. Tobacco use behaviors are reported as percentages, representing the proportion of surveyed students in that state and year who selected a specific response. For example, a value of 18 would indicate that 18 percent of surveyed students in that state reported the behavior described. Unlike our first dataset, this data was already structured and formatted, so we did not manually transcribe or recode the original source beyond cleaning and filtering for relevant variables.   

This dataset is useful because it allows us to compare youth tobacco use across states and over time within our 2000 to 2017 research period. However, there are limitations. The survey is self-administered, meaning responses rely on student honesty and may be affected by underreporting or social desirability bias. Not all states participated in every survey year, which creates gaps in longitudinal coverage. Some entries with small sample sizes were suppressed or combined, reducing precision for certain subgroups and requiring careful filtering. These factors must be considered when interpreting trends in youth tobacco use

#### Section 4 of EDA - Exploring the Relationship Between Policy Strength and Youth Vaping

This section explores the relationships between state-level vaping policies, youth vaping prevalence, and public sentiment toward vaping. After combining the policy, youth survey, and sentiment datasets into a single dataset, we use a series of visualizations and statistical analyses to examine patterns across states. The goal of this exploratory analysis is to understand how differences in policy environments may be associated with variations in youth vaping rates and public attitudes, and to identify trends that can inform the later stages of our analysis.

```python
# Convert Yes/No policy columns into numeric values
policy_eda_data['Permit_num'] = policy_eda_data['Permit Required'].map({'Yes':1,'No':0})
policy_eda_data['SmokeFree_num'] = policy_eda_data['Smoke-Free Restrictions'].map({'Yes':1,'No':0})
policy_eda_data['YouthRestrict_num'] = policy_eda_data['Youth Restrictions'].map({'Yes':1,'No':0})
policy_eda_data['Packaging_num'] = policy_eda_data['Product Packaging'].map({'Yes':1,'No':0})
policy_eda_data['Tax_num'] = policy_eda_data['Tax'].map({'Yes':1,'No':0})
# Normalize age so it contributes but does not dominate
policy_eda_data['Age_scaled'] = policy_eda_data['Age'].fillna(0) / policy_eda_data['Age'].max()
```

```python
# Force columns to numeric so math works
cols = ['Permit_num','SmokeFree_num','YouthRestrict_num','Packaging_num','Tax_num']
policy_eda_data[cols] = policy_eda_data[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
```

To simplify comparisons across states, we created a policy strength score that summarizes multiple vaping-related policies into a single numeric value. Individual policy measures can be difficult to interpret when analyzed separately, especially when states implement different combinations of regulations. By aggregating these measures into one score, we are able to represent the overall regulatory environment of each state on a consistent scale. Higher scores indicate stronger or more restrictive policy environments, allowing us to more easily compare how policy strength relates to youth vaping outcomes across states.

```python
policy_eda_data['policy_score'] = (
    policy_eda_data['Permit_num'] +
    policy_eda_data['SmokeFree_num'] +
    policy_eda_data['YouthRestrict_num'] +
    policy_eda_data['Packaging_num'] +
    policy_eda_data['Tax_num'] +
    policy_eda_data['Age_scaled'])

policy_eda_data[['StateAbbr','policy_score']].head()
```

```python
youth_state = (yts_eda_data[
        (yts_eda_data['Gender'] == 'Overall') &
        (yts_eda_data['Education'] == 'High School') &
        (yts_eda_data['Response'] == 'Current') &
        (yts_eda_data['TopicDesc'] == 'Smokeless Tobacco Use (Youth)') &
        (yts_eda_data['LocationDesc'] != 'National (States and DC)')].groupby('LocationDesc')['Data_Value'].mean().reset_index())
youth_state.columns = ['State','avg_youth_use']
```

```python
tweet_state = (tweet_eda_data.dropna(subset=['StateAbbr']).groupby('StateAbbr')['VADER_sentiment_score'].mean().reset_index())
tweet_state.columns = ['StateAbbr','avg_sentiment']
```

To statistically evaluate whether youth vaping rates differ across policy strength levels, we conducted a one-way Analysis of Variance (ANOVA). This test compares the mean youth vaping rates across the different policy score categories to determine whether the differences observed in the boxplot are statistically significant rather than occurring by chance.

```python
from scipy.stats import f_oneway

# group youth vaping rates by policy category
groups = [group["avg_youth_use"].dropna()
    for name, group in combined_data.groupby("policy_score_cat")]

# run ANOVA
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

The results of the one-way ANOVA indicate that there are statistically significant differences in average youth vaping rates across different policy strength levels (F = 4.90, p = 0.003). Since the p-value is well below the conventional significance threshold of 0.05, we reject the null hypothesis that all policy groups have the same average youth vaping rate. This suggests that the strength of vaping-related policies is associated with differences in youth vaping prevalence across states. Combined with the earlier visualizations and correlation analysis, these results provide additional evidence that stronger policy environments may be linked to lower youth vaping rates.
